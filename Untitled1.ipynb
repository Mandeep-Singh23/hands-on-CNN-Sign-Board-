```python
from sklearn.utils import shuffle
X_train, y_train = shuffle(X_train, y_train)
```

```python
X_train_grey = np.sum(X_train/3, axis = 3, keepdims = True)
X_test_grey = np.sum(X_test/3, axis = 3, keepdims = True)
X_val_grey = np.sum(X_val/3,axis = 3, keepdims = True)

print(X_train_grey.shape)
print(X_test_grey.shape)
print(X_val_grey.shape)
```

```python
X_train_norm = X_train_grey/255.0
X_test_norm = X_test_grey/255.0
X_val_norm = X_val_grey/255.0
np.max(X_train_norm),np.max(X_test_norm),np.max(X_val_norm),np.min(X_train_norm),np.min(X_test_norm),np.min(X_val_norm)
```

```python
X_train_norm_1 = (X_train_grey-255)/255
X_test_norm_1 = (X_test_grey-255)/255
X_val_norm_1 = (X_val_grey-255)/255
```

```python
np.max(X_train_norm_1),np.max(X_test_norm_1),np.max(X_val_norm_1),np.min(X_train_norm_1),np.min(X_test_norm),np.min(X_val_norm_1)
```

```python
from tensorflow.keras import models, layers, datasets
```

```python
model = models.Sequential()
model.add(layers.Conv2D(6,(5,5), activation = 'relu', input_shape = (32,32,1)))
model.add(layers.MaxPooling2D())
model.add(layers.Dropout(0.2))
model.add(layers.Conv2D(16, (5,5), activation = 'relu'))
model.add(layers.MaxPooling2D())
model.add(layers.Flatten())
model.add(layers.Dense(120, activation = 'relu'))
model.add(layers.Dense(80, activation = 'relu'))
model.add(layers.Dense(43, activation = 'softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer = 'Adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

model fit using normaliztion from -1 to 0

```python
model_fit = model.fit(X_train_norm_1, y_train, batch_size = 500, epochs = 15, verbose = 1, validation_data= (X_val_norm_1,y_val))
```

model fit using stardization ranging from 0 to 1

```python
model_fit_new = model.fit(X_train_norm, y_train, batch_size = 500, epochs = 15, verbose = 1, validation_data= (X_val_norm,y_val))
```
